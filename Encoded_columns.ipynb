```python
import pandas as pd
df = pd.read_csv('/content/telco-customer-churn.zip')
```

```python
df.head(5)
```

```python
df.info()
```

```python
text_columns = df.select_dtypes(object).columns
```

```python
text_columns.unique()
```

```python
for col in text_columns:
  print(f"{col} : {df[col].nunique()} unique values")
```

```python
binary_col=[]
three_col=[]
for col in text_columns:
  unique_values = df[col].nunique()
  if unique_values <= 2:
    binary_col.append(col)
  else:
    three_col.append(col)
```

```python
binary_col
```

```python
three_col
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
for col in binary_col:
  df[col]=le.fit_transform(df[col])
```

```python
df = pd.get_dummies(df, columns=three_col, drop_first=True)
```
